```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Load data
df = pd.read_csv('cleaned_car_data.csv')

# Preview the data
df.head()
```

# reviews / brand

```python
# Group by brand and count the number of reviews
brand_reviews = df['Brand'].value_counts()

# Plot the number of reviews per brand
plt.figure(figsize=(10,10))
sns.barplot(x=brand_reviews.values, y=brand_reviews.index, palette='viridis')
plt.title('Number of Reviews per Brand')
plt.xticks(rotation=45)
plt.ylabel('Number of Reviews')
plt.xlabel('Brand')
plt.show()
```

```python
# Group by brand and count the number of reviews
brand_reviews = df['Brand'].value_counts()

# Calculate the percentage of reviews for each brand
total_reviews = brand_reviews.sum()  # Total number of reviews
percentage_reviews = (brand_reviews / total_reviews) * 100  # Calculate percentage

# Plot the percentage of reviews per brand
plt.figure(figsize=(10, 10))
sns.barplot(x=percentage_reviews.values, y=percentage_reviews.index, palette='viridis')
plt.title('Percentage of Reviews per Brand')
plt.xticks(rotation=0)
plt.ylabel('Brand')
plt.xlabel('Percentage of Reviews (%)')
plt.xlim(0, 8)  # Set the x-axis limits from 0 to 8%

# Add grey dashed grid lines for easier reading of percentage positions
plt.grid(axis='x', color='grey', linestyle='--', linewidth=0.7)

plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Sample DataFrame creation (Remove this if you're using your own DataFrame)
# df = pd.DataFrame({
#     'Brand': ['A', 'B', 'C', 'A', 'B', 'C', 'D', 'E', 'F', 'G', 'H'],
#     'Rating': [4, 5, 3, 4, 2, 5, 3, 4, 5, 4, 3]
# })

# Group by brand and count the number of reviews
brand_reviews = df['Brand'].value_counts()

# Generate random colors for each brand
np.random.seed(58008)  # For reproducibility
colors = np.random.rand(len(brand_reviews), 3)  # Generate random RGB colors

# Calculate the percentage of reviews for each brand
total_reviews = brand_reviews.sum()  # Total number of reviews
percentage_reviews = (brand_reviews / total_reviews) * 100  # Calculate percentage

# Plot the pie chart for the number of reviews per brand
plt.figure(figsize=(8, 8))  # Set the figure size to make the pie chart larger
plt.pie(brand_reviews, colors=colors, startangle=90, 
        wedgeprops={'edgecolor': 'white'})  # Use wedgeprops to add borders

# Set the title
plt.title('Percentage of Reviews per Brand')

# Add a legend to show brand names and their corresponding percentages
plt.legend(labels=[f'{brand} - {percentage:.2f}%' for brand, percentage in zip(brand_reviews.index, percentage_reviews)],
           title="Brands", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))

plt.axis('equal')  # Equal aspect ratio ensures the pie is drawn as a circle
plt.show()
```

# Score / brand

```python
# Calculate the average rating for each brand and sort in descending order
avg_rating_per_brand = df.groupby('Brand')['Rating'].mean().sort_values(ascending=False)

# Plot bar plot of average ratings for each brand
plt.figure(figsize=(12, 8))
sns.barplot(x=avg_rating_per_brand.index, y=avg_rating_per_brand.values, palette='muted')
plt.title('Average Rating for Each Brand')
plt.xlabel('Brand')
plt.ylabel('Average Rating')
plt.xticks(rotation=90)  # Rotate brand names for better visibility
plt.ylim(0, 10)  # Set y-axis limits based on the expected rating range
plt.show()
```

# reviews / year

```python
# Group by Year and count the number of reviews
reviews_per_year = df.groupby('Year').size()

# Plot reviews per year
plt.figure(figsize=(10,6))
reviews_per_year.plot(kind='bar', color='skyblue')
plt.title('Number of Reviews per Year')
plt.xlabel('Year')
plt.ylabel('Number of Reviews')
plt.xticks(rotation=45)  # Rotate brand names for better visibility
plt.show()
```

# avg score / brand

```python
# Group by Brand and calculate average rating
avg_rating_per_brand = df.groupby('Brand')['Rating'].mean().sort_values(ascending=True)

# Plot average score per brand with horizontal bars and increased spacing
plt.figure(figsize=(10,10))  # Increase the height to add more space between bars
avg_rating_per_brand.plot(kind='barh', color='seagreen', width=0.8)  # Adjust width to control bar size
plt.title('Average Rating per Brand')
plt.xlabel('Average Rating')
plt.ylabel('Brand')
plt.xlim(5, 10)  # Set x-axis limits based on the expected rating range
plt.show()
```

# avg words / review

```python
# Combine the relevant text columns into one
df['CombinedReview'] = df[['Price Feature', 'Design Feature', 'Practicality Feature', 
                           'Under Bonnet Feature', 'Efficiency Feature', 'Driving Feature', 
                           'Safety Feature', 'Ownership Feature']].astype(str).agg(' '.join, axis=1)

# Count the number of words in the combined review
df['WordCount'] = df['CombinedReview'].apply(lambda x: len(str(x).split()))

# Group by Brand and calculate the average number of words per review
avg_words_per_review = df.groupby('Brand')['WordCount'].mean().sort_values(ascending=True)

# Plot the average number of words per review per brand
plt.figure(figsize=(10,10))
avg_words_per_review.plot(kind='barh', color='orange')
plt.title('Average Number of Words per Review for Each Brand')
plt.xlabel('Average Word Count')
plt.ylabel('Brand')
plt.xticks(rotation=45)
plt.show()

# Optionally print the average word counts to verify
print(avg_words_per_review)
```

### ...with mean overall wc

```python
# Combine the relevant text columns into one
df['CombinedReview'] = df[['Price Feature', 'Design Feature', 'Practicality Feature', 
                           'Under Bonnet Feature', 'Efficiency Feature', 'Driving Feature', 
                           'Safety Feature', 'Ownership Feature']].astype(str).agg(' '.join, axis=1)

# Count the number of words in the combined review
df['WordCount'] = df['CombinedReview'].apply(lambda x: len(str(x).split()))

# Group by Brand and calculate the average number of words per review
avg_words_per_review = df.groupby('Brand')['WordCount'].mean().sort_values(ascending=False)

# Calculate the overall mean word count
overall_mean_word_count = df['WordCount'].mean()

# Plotting the horizontal barplot of average word count per brand
plt.figure(figsize=(12, 10))
sns.barplot(y=avg_words_per_review.index, x=avg_words_per_review.values, color='orange')  # Swap x and y
plt.title('Average Word Count per Brand')
plt.xlabel('Average Word Count')
plt.ylabel('Brand')

# Adding a reference line for the overall mean word count
# The axvline() is used for a vertical line on the x-axis
# plt.axvline(overall_mean_word_count, color='red', linestyle='solid', label='Overall Mean (solid)')  # Solid line
# plt.axvline(overall_mean_word_count, color='red', linestyle='dotted', label='Overall Mean (dotted)')  # Dotted line
plt.axvline(overall_mean_word_count, color='red', linestyle='dashed', label='Mean Word Count', )  # Dashed line
# plt.axvline(overall_mean_word_count, color='red', linestyle='dashdot', label='Overall Mean (dashdot)')  # Dash-dot line

# Add a legend for the reference line
plt.legend(loc='lower right')

plt.show()
```

# Correlation Heatmap of Features

```python
corr_matrix = df[['Rating', 'Price Feature Rating', 'Design Feature Rating', 'Practicality Feature Rating', 
                  'Under Bonnet Feature Rating', 'Efficiency Feature Rating', 'Driving Feature Rating', 
                  'Safety Feature Rating', 'Ownership Feature Rating']].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Between Features and Rating')
plt.show()
```

# Top Brands by Feature Ratings

```python
feature_ratings = df.groupby('Brand')[['Design Feature Rating', 'Safety Feature Rating', 'Efficiency Feature Rating']].mean().sort_values(by='Safety Feature Rating', ascending=True)

feature_ratings.plot(kind='barh', figsize=(12, 12), colormap='tab20')
plt.title('Average Feature Ratings per Brand')
plt.ylabel('Average Feature Rating')
plt.xticks(rotation=90)
plt.show()
```
